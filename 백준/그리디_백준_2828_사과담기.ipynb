```python
screen_size, basket_size = map(int,input().split())

apple_number = int(input())

location = []

basket = []

distance = 0

for i in range(apple_number):
    
    location.append(int(input()))
        
# 바스켓 설정

for j in range(1,basket_size+1):
    
    basket.append(j)
    

    
# 바스켓의 오른쪽 끝 값보다 큰 경우

for m in location:
    
    if basket_size == 1:
        
        if basket[basket_size-1] < m:

            distance = distance + (m - basket[basket_size-1])
            
            basket[0] = basket[0] + (m - basket[0])
            
         


        elif m < basket[basket_size-1]: 

            distance = distance + (basket[basket_size-1] - m)
            
            basket[0] = basket[0] + (m - basket[0])

            
            
            
        else :

            distance = distance + 0
            
           
        
    else:
        
        if basket[basket_size-1] < m:

            distance = distance + (m - basket[basket_size-1])
            
            basket[0] = basket[0] + (m - basket[basket_size-1])
            
            basket[basket_size-1] = basket[basket_size-1] + (m - basket[basket_size-1])
            
            

        
            
        elif m <= basket[basket_size-1] and basket[0] <= m : 

            distance = distance + 0
            
            

        else :

            distance = distance + (basket[0] - m) 
            
            basket[basket_size-1] = basket[basket_size-1] + (m - basket[0])
            
            basket[0] = basket[0] + (m - basket[0])
            
        
            
    
print(distance)
```
